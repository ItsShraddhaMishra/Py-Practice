```python
import matplotlib.pyplot as plt
import numpy as np
```

### Plotting X & Y axes

```python
xpoints = np.array([0, 6])
ypoints = np.array([0, 250])
plt.plot(xpoints, ypoints)
plt.title("Plotting X-Y axes")
plt.show()
```

##### Plotting without a line

```python
x = np.array([1, 8])
y=  np.array([3, 10])
plt.plot(x, y, "o")
plt.show()
```

#### Markers

```python
y = np.array([3, 8, 1, 10])
plt.plot(y, marker = 'o')
plt.show()
```

#### Format Strings

__marker | line | color__

```python
y = np.array([3, 8, 1, 10])
plt.plot(y, 'D:g')
plt.show()
```

#### Makersize 

* ms = markersize
* mec = markercolor
* mfc = markerfacecolor

```python
y = np.array([3, 8, 1 , 10])
plt.plot(y, marker = 'o', ms=20, mec = 'b', mfc = 'r')
plt.xlabel("X-axis")
plt.ylabel('Y-axis')
plt.show()
```
